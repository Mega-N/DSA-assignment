**Visualizing The Percentage of Respondents Interested in the field of data Science Areas**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
topics = pd.DataFrame({
    'Data Science Area': ['Machine Learning', 'Data Analysis', 'Big Data', 'Data Visualization', 'Deep Learning'],
    'Very Interested': [75, 60, 45, 55, 70],
    'Somewhat Interested': [15, 20, 30, 25, 15],
    'Not Interested': [10, 20, 25, 20, 15]
})

# Create the bar chart
ax = topics.plot(kind='bar',
                 figsize=(20, 8),
                 width=0.8,
                 color=['#5cb85c', '#5bc0de', '#d9534f'])

plt.title('The Percentage of Respondents Interested in Data Science Areas', fontsize=16)
plt.legend(fontsize=14)
plt.xticks(fontsize=14)
plt.yticks([])

# Display percentages above the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate('{}%'.format(height), (x + width/2 - 0.1, y + height + 1), fontsize=14)

# Remove left, top, and right borders
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
